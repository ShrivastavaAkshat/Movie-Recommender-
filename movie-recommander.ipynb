```python
def changeGenres(genresList):
    new_list = []
    for i in ast.literal_eval(genresList):
        new_list.append(i['name'])
    return new_list
```

```python
movies['genres'] = movies['genres'].apply(changeGenres)
```

```python
movies['keywords'] = movies['keywords'].apply(changeGenres)
```

```python
def getCast(castList):
    count = 0
    new_list = []
    for i in ast.literal_eval(castList):
        count = count+1
        if count <= 5:
            new_list.append(i['name'])
        else:
            break
    return new_list
```

```python
movies['cast'] = movies['cast'].apply(getCast)
```

```python
def getDirector(crewdetails):
    job = "Director"
    new_list = []
    for i in ast.literal_eval(crewdetails):
        if i['job'] == job:
            new_list.append(i['name'])
    return new_list
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# apply stem from nltk librabry to rmove plural words or words with same meaning
```

```python
stemmer = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(stemmer.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Example of stemeing 
stem('actions')
```

```python
# convert text to vector with bags of word technique a library of scikit learn
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
new_df.head()
```

```python
cv.get_feature_names()
```

```python
# Find the cosine distance between two vectors, the smaller the angle or cosine distance the closer the vectors are
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
def recommed_movies(moviename):
    movie_index = new_df[new_df['title'] == moviename].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
    
#     print(movie_list)
    
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommed_movies('Code 46')
```
